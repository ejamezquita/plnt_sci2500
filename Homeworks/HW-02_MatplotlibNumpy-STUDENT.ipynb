## 1. Vogel's sunflowers (13 pts)

Remember from the last homework that the spirals in a plant (phyllotaxis) follow either Fibonacci or a Fibonacci-like numbers. Behind all these Fibonacci numbers, is the **golden ratio**:
$$\varphi = \frac{1+\sqrt5}2.$$ 

A $\varphi$-th of a circumference is $360^\circ/\varphi \approx 222.5^\circ$. Its complement is the **golden angle** 

$$g  = 360^\circ - 360^\circ/\varphi \approx 137.5^\circ \approx 2.4\:\text{rad}.$$
